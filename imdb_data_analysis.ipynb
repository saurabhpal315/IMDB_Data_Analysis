```python
##what are the three most popular movies
curr.execute('select title,popularity from movies order by(popularity) desc limit 3')
curr.fetchall()
```

```python
##what is the most awarded average movie rated since jan 1st 2000
curr.execute("select title from movies where release_date>'2000-01-01' order by(vote_average) desc limit 1")
curr.fetchall()
```

```python
##which movieds were directed by Brenda Chapman?
curr.execute("select title from movies where director_id=(select id from directors where name='Brenda Chapman')")
curr.fetchall()
```

```python
##or
curr.execute("select m.title from movies as m join directors as d on m.director_id=d.id where d.name='Brenda Chapman'")
curr.fetchall()
```

```python
##name of the director who has made most number of movies
curr.execute("select count(m.title),d.name from movies as m join directors as d on m.director_id=d.id group by(d.name) order by count(m.title) desc limit 1  ")
curr.fetchall()
```

```python
##name of the most bankable directors
curr.execute("select d.name,sum(m.budget) from movies as m join directors as d on m.director_id=d.id group by(d.name) order by(sum(m.budget)) desc limit 1")
curr.fetchall()
```

```python
##tell everything about top 10 highest budget making movies
curr.execute("select *from movies order by budget desc limit 10 ")
pd.DataFrame(curr.fetchall())
```

```python
##revenur analysis
##find top 10 revenue making movies
curr.execute('select title from movies order by revenue desc limit 10')
pd.DataFrame(curr.fetchall())
```

```python
##name all the directors with the number of movies and revenue
curr.execute("select d.name,count(m.title) as num ,sum(m.revenue) as total from movies as m join directors as d where m.director_id=d.id group by name order by sum(m.revenue) desc limit 5")
pd.DataFrame(curr.fetchall(),columns=['name','number of movies','total revenue'])
```

```python
##how many male directors have generated revenue more than 3227483234
curr.execute("select count(*) from (select d.name,count(d.name),sum(m.revenue) from movies as m join directors as d on m.director_id=d.id group by name having sum(m.revenue)>3227483234 and gender=2)")
pd.DataFrame(curr.fetchall(),columns=['Count'])
```

```python
curr.execute("select d.name,count(d.name),sum(m.revenue) from movies as m join directors as d on m.director_id=d.id group by name having sum(m.revenue)>3227483234 and gender=2")
pd.DataFrame(curr.fetchall(),columns=['Names','Number of movies','Revenue'])
```
